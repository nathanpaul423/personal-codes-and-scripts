```python
import pandas as pd
import os
import re
```

```python
target_dir = r'C:\Users\NATHANPAULBUSTAMANTE\OneDrive\Desktop\Personality test\Fuego\Weedmaps - View vs Clicks\2023'
```

```python
file_list =os.listdir(target_dir)
```

```python
file_list
```

```python
date_pattern = r'\d{2}-\d{2}-\d{4}'
```

```python
dates = [re.search(date_pattern, filename).group() for filename in file_list]
```

```python
dates
```

```python
columns = ['Listing Name', 'Listing Region', 'Listing Type', 'Prorated Spend',
       'Premium Days', 'Total Clicks', 'Deal Claims', 'Deal Views',
       'Cost Per Click', 'Listing Page Visitors', 'Cost Per LPV',
       'Contact Clicks', 'Phone Clicks', 'Direction Clicks', 'Website Clicks',
       'Email Clicks', 'Order Count', 'Order GMV', 'AOV', 'eROAS','date']
```

```python
compilation = pd.DataFrame(columns=columns)
```

```python
for i, (x, y)  in enumerate(zip(file_list,dates)):
    df1 = pd.read_csv(r'C:\Users\NATHANPAULBUSTAMANTE\OneDrive\Desktop\Personality test\Fuego\Weedmaps - View vs Clicks\2023\\'+ x )
    df1['date'] = y

    compilation = compilation.append(df1,ignore_index = True)
```

```python
compilation
```

```python
compilation.to_csv(r'C:\Users\NATHANPAULBUSTAMANTE\OneDrive\Desktop\Personality test\Fuego\Weedmaps - View vs Clicks\All Listings - Historical Products SeptOct2023.csv')
```
